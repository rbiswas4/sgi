This is a notebook with a toy model illustrating the setup of SNe Catalogs. The essential ingredients are 
- A large number of samples of objects with a parameter 'x' distributed with a known distribution. To think of current plans, these objects are analogous to galaxies, and this parameter 'x' is analogous to the stellar mass of the galaxy.
- For definiteness, we take the distribution of the 'x' to be exponential. For galaxies, this may be described by function close to a Schechter function, or even better by histograms of stellar masses of catalog galaxies. 
- For each object, we can write down the probability of some event given the value x P(event|x). Our objective is to obtain samples of the event which have to 
    1. respect the distribution ie. (normalized histograms should match the mathematical experession)
    2. The number of samples should match a prescribed number of samples. This cannot be done for numevents too high in the current method as that may imply probability greater than 1 on an object. We may have to introduce a second distribution to split probabilities on an event

```python
import numpy as np
```

```python
np.random.seed(0)
datasize=int(1e6)
numEvents  = int(5.e3)
# data = np.random.uniform(0., 1., size=datasize)
data = np.random.exponential(scale= 1., size=datasize)
```

```python
randoms = np.random.uniform(0., 1., size=len(data))
# k = data.max()
k = numEvents / data.sum()
mask = randoms / k  / data  < 1.0
# This mask2 is just to show that one can get the distribution without using
# numEvents if the number is not specified, and can thus use it for oversampling
mask2 = randoms / k / data < 5.0
```

### Conditional Distribution

```python
bincents = 0.5 * (bins[1:] + bins[:-1])
# plt.plot(bincents, norigu)
mydf = pd.DataFrame({'bincents':bincents, 'Pcond':ncond/norig})
mydf['binInds']  = mydf.bincents // 0.5
largerBinsdf = mydf.groupby('binInds').mean()
largerBinsdf['PcondErr'] = mydf.groupby('binInds').Pcond.std()
plt.errorbar(largerBinsdf.bincents, largerBinsdf.Pcond, yerr=largerBinsdf.PcondErr,
             fmt='o', label='Samples')
plt.plot(bincents, bincents, '--r', label='Correct expression: x')
plt.title('The Conditional Distribution in bins of 0.5')
plt.xlabel('x')
plt.ylabel('P(even|x) from Samples')
plt.legend(loc='best')
```

### 'Time Grids'

In the above methods, while we are doing this pass of selecting objects, we forget about objects that are not selected. In some fields, it may be useful to keep all objects indexed in a way that such masks can be easily
created from the indices. So, this is more functionality (for example for oversampling/adding events from other surveys) at the cost of more disk space for indices. 

```python
np.random.seed(42)
widths = 0.5 / k / data
t = np.array(map(lambda x: np.random.uniform(-x, x), widths))
```

```python
fullData = pd.DataFrame({'x':data, 't': t })
```

```python
mask = (fullData.t > -0.5) & (fullData.t < 0.5)
```
