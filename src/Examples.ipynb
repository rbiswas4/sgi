This is a notebook with a toy model illustrating the setup of SNe Catalogs. The essential ingredients are 
- A large number of samples of objects with a parameter 'x' distributed with a known distribution. To think of current plans, these objects are analogous to galaxies, and this parameter 'x' is analogous to the stellar mass of the galaxy.
- For definiteness, we take the distribution of the 'x' to be exponential. For galaxies, this may be described by function close to a Schechter function, or even better by histograms of stellar masses of catalog galaxies. 
- For each object, we can write down the probability of some event given the value x P(event|x). Our objective is to obtain samples of the event

```python
import numpy as np
```

```python
datasize=int(1e6)
data = np.random.uniform(0., 1., size=datasize)
data = np.random.exponential(scale= 1., size=datasize)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
sns.set_style('whitegrid')
```

```python
bins = np.linspace(0., 10., 1000.)
```

```python
randoms = np.random.uniform(0., 1., size=len(data))
k = data.max()
mask = randoms * k  / data  < 1.0
mask2 = randoms * k / data < 5.0
```

```python
norig, bins, patches = plt.hist(data, bins=bins, histtype='step', normed=1, lw=2, label='expon samples')
rets = plt.hist(data, weights=data, bins=bins, histtype='step', normed=1, lw=8, color='g', label='weighted samples')

ncond2, bins, patches = plt.hist(data[mask2], bins=bins, histtype='step', normed=1, color='r', lw=1., label='more samples')
ncond, bins, patches = plt.hist(data[mask], bins=bins, histtype='step', normed=1, color='b', lw=1., label='less samples')
plt.axvline(1)
x = np.arange(0, 10, 0.1)
plt.plot(x, np.exp(-x),'--k',lw=2., label='Exp-plot')
plt.plot(x, x * np.exp(-x),'--k',lw=4., label='x*exp(-x)')
#plt.plot(x,2. *x)
plt.legend(loc='best')
plt.ylabel('PDF')
plt.xlabel('x')
plt.title('Normalized Distributions of x*exp(-x)')
```

```python
norig, bins, patches = plt.hist(data, bins=bins, histtype='step', normed=0, lw=2, label='exponential')
ncond, bins, patches = plt.hist(data[mask], bins=bins, histtype='step', normed=0, color='k', lw=1., label='data.max()')
ncond2, bins, patches = plt.hist(data[mask2], bins=bins, histtype='step', normed=0, color='r', lw=1.,label='data.max() /5')



rets = plt.hist(data, weights=data, bins=bins, histtype='step', normed=0, lw=3, color='g', label='weights')
plt.legend(loc='best')
plt.xlabel('x')
plt.ylabel('Number of Samples')
```
